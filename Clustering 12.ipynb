```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Load the dataset
df = pd.read_csv("iris.csv")
```

```python
# Drop 'Id' and 'Species' (we're only using numeric features)
X = df.drop(['Id', 'Species'], axis=1)
```

```python
# Apply KMeans with 3 clusters (since there are 3 Iris species)
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
```

```python
# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_
```

```python

# Plot two features to visualize clusters
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c=df['Cluster'], cmap='viridis')
plt.title('KMeans Clustering (Iris Dataset)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(True)
plt.show()
```
