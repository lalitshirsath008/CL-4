```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
data = pd.read_csv("sentiment_data_1000.csv")
```

```python
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['sentiment'], test_size=0.2, random_state=42)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=10000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
```

```python
y_pred = model.predict(X_test_tfidf)
```

```python

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
text_to_analyze = ["This restaurant is amazing! The food was delicious."]
text_to_analyze_tfidf = tfidf_vectorizer.transform(text_to_analyze)
predicted_sentiment = model.predict(text_to_analyze_tfidf)
```

```python
print("Predicted sentiment:", predicted_sentiment)
```
